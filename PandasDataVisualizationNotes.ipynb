```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

np.random.seed(0)
df = pd.DataFrame({
    "A": np.random.randn(100),
    "B": np.random.randn(100) * 2,
    "C": np.random.randint(1, 50, 100),
    "D": np.random.rand(100)
})
```

```python
# hist

df["A"].hist()
```

```python
df["A"].hist(bins=30)
```

```python
df.plot(kind="hist")
```

```python
df.plot.hist()
```

```python
# area

df.plot.area(alpha=0.5, stacked=False)
```

```python
# bar

df.head().plot.bar()
```

```python
df.head().plot.bar(stacked=True)
```

```python
# line

df.plot.line(x="A", y="B", lw=0.5, figsize=(9,3))
```

```python
# scatter

df.plot.scatter(x="A", y="B", c="C", cmap="coolwarm", s=df["C"]*10)
```

```python
# box

df.plot.box()
```

```python
# hexbin

df.plot.hexbin(x="A", y="B", gridsize=15, cmap="coolwarm")
```

```python
# kde

df.plot.kde()
```

```python
# density

df.plot.density()
```
