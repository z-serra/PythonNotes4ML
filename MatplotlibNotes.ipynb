```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0, 5, 10)
y = x ** 2

# plot

plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('X and Y')
```

```python
# subplot

plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')
plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

```python
# figure

fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes.plot(x, y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('TITLE')
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.9, 0.9])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.4])

axes1.plot(x, y, 'r', label='Label 1')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('X vs. Y')
axes1.legend(loc=4)

axes2.plot(y, x, 'b', label='Label 2')
axes2.set_xlabel('Y')
axes2.set_ylabel('X')
axes2.set_title('Y vs. X')
axes2.legend(loc=0)
```

```python
# subplots
fig, axes = plt.subplots(nrows=2, ncols=2)
for current_ax in axes.flat:
    current_ax.plot(x, y)
plt.tight_layout()
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x, y)
axes[0].set_title('1st Plot')
axes[1].plot(y, x)
axes[1].set_title('2nd Plot')
plt.tight_layout()
```

```python
fig = plt.figure(figsize=(2,2), dpi=100) # dots/pixels per inch
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(2,2))
axes[0].plot(x, y)
axes[1].plot(y, x)
plt.tight_layout()
```

```python
# savefig

fig.savefig('my_fig.png', dpi=200)
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# lw = line width, ls = line style

ax.plot(x, y, color='pink', lw=10, alpha=0.5, ls='--', marker='+', markersize=20,
        markeredgecolor='red', markeredgewidth=10)

ax.set_xlim([0, 5])
ax.set_ylim([0, 25])
```

```python
# scatter

plt.scatter(x, y)
```

```python
# hist

from random import sample
data = sample(range(1, 1000), 100)
plt.hist(data)
```

```python
# boxplot

data = [np.random.normal(0, std, 100) for std in range(1, 4)]
plt.boxplot(data, vert=True, patch_artist=True)
```
