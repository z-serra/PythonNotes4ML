```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

import plotly.express as px
```

```python
df = pd.DataFrame(np.random.randn(100, 4), columns='A B C D'.split())
df2 = pd.DataFrame({'Category': ['A', 'B', 'C'], 'Values': [30, 40, 50]})
```

```python
df.plot()
```

```python
# line

px.line(df)
```

```python
# scatter

px.scatter(df, x="A", y="B")
```

```python
# bar

px.bar(df2, x="Category", y="Values")
```

```python
px.bar(df)
```

```python
px.bar(df.count())
```

```python
px.bar(df.sum())
```

```python
# box

px.box(df)
```

```python
import plotly.graph_objects as go

go.Figure(data=[go.Surface(z=[[1, 2, 3, 4, 5], [10, 20, 30, 20, 10], [500, 400, 300, 200, 100]])])
```

```python
go.Figure(data=[go.Surface(z=[[1, 2, 3, 4, 5], [10, 20, 30, 20, 10], [5, 4, 3, 2, 1]], colorscale="rdylbu")])
```

```python
# histogram

px.histogram(df)
```

```python
px.histogram(df["A"], nbins=50)
```

```python
# violin

px.violin(df, y=["A","B"])
```

```python
# scatter matrix

px.scatter_matrix(df)
```
