```python
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

tips.head()
```

```python
# displot

sns.displot(tips['total_bill'], kde=True, bins=30) # kde = kernel density estimation
```

```python
# kdeplot
sns.kdeplot(tips['total_bill'])
```

```python
# jointplot

sns.jointplot(data=tips, x='total_bill', y='tip')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='hex')
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='reg') # regression
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip', kind='kde')
```

```python
# pairplot

sns.pairplot(tips, hue='sex', palette='coolwarm')
```

```python
# rugplot

sns.rugplot(tips['total_bill'])
# normal distributions centered around dashes
# sum of all normal distributions = kde plot
```

```python
# barplot

sns.barplot(data=tips, x='sex', y='total_bill')
```

```python
sns.barplot(data=tips, x='sex', y='total_bill', estimator=np.std)
```

```python
# countplot

sns.countplot(data=tips, x='sex')
```

```python
# boxplot

sns.boxplot(data=tips, x='day', y='total_bill')
```

```python
sns.boxplot(data=tips, x='day', y='total_bill', hue='smoker')
```

```python
# violinplot

sns.violinplot(data=tips, x='day', y='total_bill')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill', hue='sex', split=True)
```

```python
# stripplot

sns.stripplot(data=tips, x='day', y='total_bill', jitter=True, hue='sex', dodge=True)
```

```python
# swarmplot

sns.swarmplot(data=tips, x='day', y='total_bill')
```

```python
sns.violinplot(data=tips, x='day', y='total_bill')
sns.swarmplot(data=tips, x='day', y='total_bill', color='black')
```

```python
# catplot

sns.catplot(data=tips, x='day', y='total_bill', kind='bar')
```

```python
sns.catplot(data=tips, x='day', y='total_bill', kind='violin')
```

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
# heatmap

sns.heatmap(tips.corr(numeric_only=True), annot=True, cmap='coolwarm')
```

```python
fpt = flights.pivot_table(index='month', columns='year', values='passengers')
sns.heatmap(fpt, cmap='magma', linecolor='white', linewidths=0.5)
```

```python
# clustermap

sns.clustermap(fpt, cmap='coolwarm', standard_scale=1)
```

```python
iris = sns.load_dataset('iris')
iris.head()
```

```python
sns.pairplot(iris)
```

```python
# PairGrid

sns.PairGrid(iris).map(plt.scatter)
```

```python
pg = sns.PairGrid(iris)

pg.map_diag(sns.kdeplot)
pg.map_lower(sns.kdeplot)
pg.map_upper(plt.scatter)
```

```python
# FacetGrid

sns.FacetGrid(data=tips, col='time', row='smoker').map(sns.kdeplot, 'total_bill')
```

```python
sns.FacetGrid(data=tips, col='time', row='smoker').map(plt.scatter, 'total_bill', 'tip')
```

```python
# lmplot
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', markers=['o', 'v'], scatter_kws={'s': 80})
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, col='day', row='time', hue='sex', aspect=0.5)
```

```python
plt.figure(figsize=(12,3))
#sns.set_context('poster', font_scale=2)
#sns.set_context('notebook')
sns.set_style('ticks')
sns.countplot(x='sex', data=tips)
sns.despine(left=True, bottom=True)
```
